```python
, classification_report
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier as destre
from sklearn.tree import plot_tree
from sklearn.metrics import accuracy_score,precision_score, confusion_matrix,classification_report
datos=pd.read_csv('BlueOrRed.csv')
datos
y=datos['Undecided']
x=datos.drop(['Undecided','Voter'], axis=1)
y.replace(['Yes', 'No'],[1,0],inplace=True)
arbol= destre(criterion='entropy',random_state=42, max_depth=5)
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=42,shuffle=True)
arbol=arbol.fit(x_train,y_train)
prediction= arbol.predict(x_test)
print(accuracy_score(y_test, prediction))
print(precision_score(y_test, prediction))
cf=confusion_matrix(y_test, prediction)
print(cf)
print(classification_report(y_test,prediction))
print(x.columns)
plt.figure(figsize=(14,10))
plot_tree( decision_tree=arbol ,feature_names=['Age', 'HomeOwner', 'Female', 'Married', 'HouseholdSize', 'Income','Education', 'Church'], fontsize=8)
```
